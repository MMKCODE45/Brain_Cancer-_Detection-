```python
# Load necessary modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import os
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, learning_curve
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.decomposition import PCA
```

```python
# Path to your data folder in Google Drive
data_folder = '/content/drive/MyDrive/'

# List the contents of the 'Training/' directory
path = os.listdir(data_folder + 'brain_tumor/Training/')
```

```python
# classes dictionary with all cancer classes
classes = {'no_tumor': 0, 'pituitary_tumor': 1, 'meningioma_tumor': 2, 'glioma_tumor': 3}


# Initialize lists to store image data (X) and corresponding labels (Y)
X = []
Y = []
```

```python
# Loop through each class and image, resize, and append to X and Y
for cls in classes:
    pth = data_folder + 'brain_tumor3/Training/' + cls
    for j in os.listdir(pth):
        img = cv2.imread(pth + '/' + j, 0)
        img = cv2.resize(img, (200, 200))
        X.append(img)
        Y.append(classes[cls])

# Convert lists to NumPy arrays
X = np.array(X)
Y = np.array(Y)
```

```python
# Reshape image data for further processing
X_updated = X.reshape(len(X), -1)

# Display unique labels and their counts
np.unique(Y)
pd.Series(Y).value_counts()

# Display shapes of the original and reshaped data
X.shape, X_updated.shape
```

```python
# Visualize a sample image
plt.imshow(X[0], cmap='gray')
plt.show()
```

```python
# Prepare data by reshaping
X_updated = X.reshape(len(X), -1)
X_updated.shape

# Split data into training and testing sets
xtrain, xtest, ytrain, ytest = train_test_split(X_updated, Y, random_state=10, test_size=.20)
xtrain.shape, xtest.shape
```

```python
# Check the range of pixel values in training and testing data
print(xtrain.max(), xtrain.min())
print(xtest.max(), xtest.min())

# Perform feature scaling (normalize pixel values)
xtrain = xtrain / 255
xtest = xtest / 255
```

```python
# Verify the new range of pixel values
print(xtrain.max(), xtrain.min())
print(xtest.max(), xtest.min())

# Feature Selection using PCA
from sklearn.decomposition import PCA
print(xtrain.shape, xtest.shape)
```

```python
# Apply PCA to retain 98% of the variance
pca = PCA(.98)
pca_train = xtrain
pca_test = xtest

# Display shapes of the data after PCA
pca_train.shape, pca_test.shape
```

```python
# Train models (Logistic Regression and SVM)
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
import warnings
warnings.filterwarnings('ignore')

lg = LogisticRegression(C=0.1)
lg.fit(xtrain, ytrain)

sv = SVC()
sv.fit(xtrain, ytrain)

rf = RandomForestClassifier(n_estimators=100, random_state=10)
rf.fit(xtrain, ytrain)

#rf_pca = RandomForestClassifier(n_estimators=100, random_state=10)
#rf_pca.fit(pca_train, ytrain)
```

```python
# Calculate and print accuracy for each model
def print_accuracy(model, x, y, label):
    accuracy = model.score(x, y)
    print(f"{label} - Training Accuracy: {accuracy:.4f}")
```

```python
# Logistic Regression
print_accuracy(lg, xtrain, ytrain, "Logistic Regression - Training")
print_accuracy(lg, xtest, ytest, "Logistic Regression - Testing")

# Support Vector Machine
print_accuracy(sv, xtrain, ytrain, "SVM - Training")
print_accuracy(sv, xtest, ytest, "SVM - Testing")

# Random Forest
print_accuracy(rf, xtrain, ytrain, "Random Forest - Training")
print_accuracy(rf, xtest, ytest, "Random Forest - Testing")
```

```python
# Make predictions using the SVM model on the testing set
pred = sv.predict(xtest)

# Identify misclassified samples
misclassified = np.where(ytest != pred)
misclassified
```

```python
# Display the total number of misclassified samples and an example
print("Total Misclassified Samples: ", len(misclassified[0]))
print("Example Prediction:", pred[misclassified[0][0]], "Actual Label:", ytest[misclassified[0][0]])

```

```python
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize
from sklearn.multiclass import OneVsRestClassifier

# Use decision function scores for ROC analysis
decision_scores = sv.decision_function(xtest)

# Binarize the labels
ytest_bin = label_binarize(ytest, classes=[0, 1, 2, 3])

# Compute ROC curve and ROC area for each class
fpr = dict()
tpr = dict()
roc_auc = dict()

# Loop through each class
for i in range(4):  # Assuming you have 4 classes
    # Compute ROC curve
    fpr[i], tpr[i], _ = roc_curve(ytest_bin[:, i], decision_scores[:, i])

    # Compute AUC score
    roc_auc[i] = auc(fpr[i], tpr[i])
```

```python
# Plot ROC curve for each class with cancer types as labels
plt.figure(figsize=(12, 8))
for i in range(4):
    cancer_type = {
        0: 'No Tumor',
        1: 'Pituitary Tumor',
        2: 'Meningioma Tumor',
        3: 'Glioma Tumor'
    }
    plt.plot(fpr[i], tpr[i], label=f'{cancer_type[i]} (AUC = {roc_auc[i]:.2f}')

# Plot a diagonal line for random predictions
plt.plot([0, 1], [0, 1], linestyle='--', color='gray', label='Random')

# Set plot labels and title
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Multiclass ROC Curve')

# Display legend in the lower right corner
plt.legend(loc="lower right")

# Show the plot
plt.show()
```
